```python

import csv
import math


#star class takes name, x, y, z, has attribute is visited 0,1
class star:
    def __init__(self,name,x,y,z):
        self.name = name
        self.x = float(x)
        self.y = float(y)
        self.z = float(z)
        self.visited = 0
        
    def __str__(self):
        return f"{self.name}({self.x})({self.y})({self.z})({self.visited})" #for printing out during testing
    

        
#funciton to find distance between two stars
def distanceCalc(star1,star2):
        d = math.sqrt(((star2.x - star1.x)**2) + ((star2.y - star1.y)**2) + ((star2.z - star1.z)**2))
        return d
    


 

input_file = csv.DictReader(open('hygxyz.csv'))
#declaring variables
distance = 10
starList = []
distList = []
travelList = []
totalDist = 0


#get data from hygxyz.csv
#append to starList and create star based on data
for row in input_file:
    if float(row["Distance"]) <= float(distance) and row["ProperName"] != "Sol":
        if row["ProperName"] != '':
            starList.append(star(row["ProperName"],row["X"],row["Y"],row["Z"]))
        elif row["Gliese"] != '':
            starList.append(star(row["Gliese"],row["X"],row["Y"],row["Z"]))
        elif row["BayerFlamsteed"]:
            starList.append(star(row["BayerFlamsteed"],row["X"],row["Y"],row["Z"]))
        else:
            starList.append(star(row["StarID"],row["X"],row["Y"],row["Z"]))



#set starting Star
originStar = star("Sol",0,0,0)


#known value for radius 10 can also change to len of list for other implementations 
for i in range(0,327):
    for item in starList:
        if originStar != item and item.visited == 0:
            distList.append(distanceCalc(item,originStar))
    if(item != originStar):
        smallest = min(distList)
        totalDist = totalDist + smallest
        print(originStar.name + " -> " + starList[(distList.index(smallest))].name + " : distance = " + str(round(smallest,2))
             + ", total distance = " +str(round(totalDist,2)))
        originStar = starList[(distList.index(smallest))]

        originStar.visited = 1
        starList.remove(originStar)
        distList.clear()



```
